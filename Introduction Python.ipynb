```
def primo(n):
  if n==1 or n==2:
    print(n, "es primo")
  for i in range(2,n):
    if n%i==0:
      print(n, "no es primo")
      break
    elif i==n-1:  
      print(n, "es primo")
      
demo=map(primo,range(12))
d=list(demo)
```

4- Definir una función que determine el n-ésimo término de la sucesión de Fibonacci:

```
from functools import reduce

def fibo(n):
  a_i=1
  b_i=a_i
  a_i_1=0
  for i in range(0,n):
    b_i=a_i
    a_i=a_i+a_i_1
    a_i_1=b_i
  return a_i

seq=map(fibo,range(15)) 
s=list(seq)
print(s)
```

7-Calcular la suma de los primeros n términos de la sucesión formada por las derivadas de la función x**N

```
import math

n=5 #número de derivadas
N=4 #exponente
x=5 #punto en el que evaluamos la suma

def derfun(N,n):
  if n>N:
    return 0
  else:
    return (math.factorial(N)/math.factorial(N-n))*(x**(N-n))

def sums(N,n):
  s=0
  for i in range(n):
    s=s+derfun(N,i)
  return s

print(sums(N,n))
```

8-Implementar una función que permita calcular la raíz n-ésima de un número. Para ello implementar un algoritmo iterativo que use la siguiente fórmula

```
import math as m
def raizn(x,n,maxIter,tol):
    Iter = 0
    err = tol + 1
    ra = 1
    while Iter < maxIter and err > tol:
        r = 0.5*(x/m.pow(ra,n-1) + ra)
        err = m.fabs(r-ra)
        Iter+=1
        ra = r
    return r, err, Iter 

x=4 #número real positivo
n=2 #número natural, orden de la raíz
maxIter=100 #número máximo de iteraciones del algoritmo (criterio de paro)
tol=0.01 #tolerancia de error para el criterio de paro

r2, err, Iter = raizn(x,2,maxIter,tol) 
r3, err, Iter = raizn(x,3,maxIter,tol) 
print('Raiz cuadrada de {0:4.2f} es {1:4.4f} '.format(x,r2))
print('Raiz cubica de {0:4.2f} es {1:4.4f} '.format(x,r3))
```
